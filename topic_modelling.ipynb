```python
import os
import pandas as pd
import nltk
from nltk.corpus import stopwords
from top2vec import Top2Vec
nltk.download('stopwords')

folder_name = 'data'
file_name = 'final_reviews.csv'

# Get the current working directory
current_dir = os.getcwd()

# Construct the full path to the file
file_path = os.path.join(current_dir, folder_name, file_name)

reviews = pd.read_csv(file_path)
reviews = reviews[[col for col in reviews.columns if 'Unnamed' not in col]]

# Assuming you have your data in a pandas series
# For example: df['reviews'] is the column that holds the text data
text_data = reviews['review'].sample(1000,random_state=42)

stop_words = set(stopwords.words('english'))

documents = text_data.tolist()

documents = [str(doc) for doc in documents]

def remove_stopwords(text):
    words = text.split()  # Split the document into words
    filtered_words = [word for word in words if word.lower() not in stop_words]  # Remove stopwords
    return ' '.join(filtered_words)  # Rejoin the filtered words into a cleaned document

# Remove stopwords from all documents
documents_cleaned = [remove_stopwords(doc) for doc in documents]

# Convert the pandas Series to a list of strings (Top2Vec requires a list of text documents)


# Initialize and train the Top2Vec model
# You can choose the embedding model: "universal-sentence-encoder" or "doc2vec" (default)
model = Top2Vec(documents_cleaned, speed="deep-learn", workers=4, embedding_model='doc2vec')  # Use "learn" for faster but less accurate training

# Print the number of topics found
print(f"Number of topics found: {model.get_num_topics()}")
```

```python
# Get the top N topics (for example, top 5)
num_topics = model.get_num_topics()
topic_words, scores, topic_nums = model.get_topics(num_topics=num_topics)

# Print out the top topics and their keywords
for i, topic in enumerate(topic_words):
    print(f"Topic {i+1}:")
    print(", ".join(topic))
    print()
```
